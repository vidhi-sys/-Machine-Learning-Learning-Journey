```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
print(housing.DESCR) # Print the dataset description
```

```python
#learning of model[FITTING(x,y)]
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
model = KNeighborsRegressor()
model.fit(housing.data, housing.target)
```

```python
preed=model.predict(housing.data)
```

```python
import matplotlib.pyplot as plt
plt.scatter(housing.target,preed)
```

so scattered may be needs pre-processing
data->scaling->knn->maybe predicts diffrently

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
pipe=Pipeline([("scale",StandardScaler()),("model",KNeighborsRegressor(n_neighbors=10))])
pipe.fit(housing.data,housing.target)
pred2=pipe.predict(housing.data)
plt.scatter(housing.target,pred2)
```

but we dont need to predit our dataset from one of the labels right it will be always correct in those cases
cross-validation is imprtant through gridsearch CV

```python
from sklearn.model_selection import GridSearchCV
import pandas as pd
#pipe.get_params()
mod=GridSearchCV(estimator=pipe,param_grid={"model__n_neighbors":range(1,100)},cv=5)#cv=crossvalidation
mod.fit(housing.data,housing.target);
pd.DataFrame(mod.cv_results_)
```
